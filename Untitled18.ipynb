```python
import pandas as pd
import numpy as np

# Define the number of students
num_students = 10

# Generate random grades between 0 and 20
grades = np.random.randint(0, 21, num_students)

# Create a list of student names with 10 great chemists
student_names = [
    'Marie Curie', 'Dmitri Mendeleev', 'Antoine Lavoisier', 'Rosalind Franklin',
    'Linus Pauling', 'Dorothy Hodgkin', 'Robert Boyle', 'John Dalton',
    'Jöns Jacob Berzelius', 'Michael Faraday'
]

# Create the DataFrame
students_df = pd.DataFrame({'Nombre Completo': student_names, 'Nota': grades})

# Separate 'Nombre Completo' into 'Nombre' and 'Apellido'
students_df[['Nombre', 'Apellido']] = students_df['Nombre Completo'].str.split(n=1, expand=True)

# Drop the original 'Nombre Completo' column
students_df = students_df.drop(columns=['Nombre Completo'])

# Display the DataFrame
display(students_df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming students_df is already loaded and contains the 'Nota' column

grades = students_df['Nota']

# Define the bins
bins = np.arange(0, 21, 1) # Bins from 0 to 20

# Calculate histogram data
n, bins = np.histogram(grades, bins=bins)

# Define colors based on grade ranges
colors = []
for bin_edge in bins[:-1]:
    if bin_edge < 14:
        colors.append('orange')
    elif 14 <= bin_edge <= 17:
        colors.append('gray')
    else:
        colors.append('green')

# Create the histogram using plt.bar for better color control per bin
plt.figure(figsize=(10, 6))
plt.bar(bins[:-1], n, width=1, color=colors, align='edge', edgecolor='black')


plt.xlabel('Nota')
plt.ylabel('Frecuencia')
plt.title('Histograma de Notas con Colores por Rango')
plt.xticks(np.arange(0, 21, 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
def color_grades_and_names(row):
    styles = [''] * len(row)
    if row['Nota'] < 14:
        # Color name in orange
        styles[row.index.get_loc('Nombre')] = 'color: orange'
    elif 14 <= row['Nota'] <= 17:
        # Color grade in gray
        styles[row.index.get_loc('Nota')] = 'color: gray'
    elif row['Nota'] > 17:
        # Color grade in green
        styles[row.index.get_loc('Nota')] = 'color: green'
    return styles

styled_df = students_df.style.apply(color_grades_and_names, axis=1)
display(styled_df)
```

```python
average_grade = students_df['Nota'].mean()
print(f"\nNota promedio del curso: {average_grade:.2f}")
```
